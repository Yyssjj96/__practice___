t-분포의 특징
- 모양: t-분포는 평균이 0이고, 대칭적인 형태를 가집니다. 꼬리가 두꺼워 정규분포보다 극단적인 값이 나올 확률이 높습니다.
- 자유도(degrees of freedom, df): t-분포의 형태는 자유도에 의해 결정됩니다. 자유도가 증가할수록 t-분포는 정규분포에 가까워집니다. 자유도는 일반적으로 표본 크기에서 1을 뺀 값입니다 (df = n - 1).
- 사용 시기: t-분포는 표본 크기가 작고(일반적으로 n < 30), 모집단의 표준편차를 모르는 경우에 사용합니다. 모집단의 표준편차를 모를 때는 표본 표준편차를 사용합니다.

t 분포의 활용
- 모집단 평균에 대한 신뢰구간 추정: 표본 평균을 사용하여 모집단 평균의 신뢰구간을 계산할 때.
- 모집단 평균에 대한 가설 검정: 예를 들어, 표본 평균이 특정 값과 다른지 여부를 검정할 때.

```python
import numpy as np
import scipy.stats as stats

# 샘플 데이터
data = [12, 15, 14, 10, 13, 15, 14, 13, 14, 12, 11, 15, 13, 13, 14]

# 샘플 평균과 표준편차
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1은 표본 표준편차를 계산하기 위함
n = len(data)

# 신뢰수준 설정 (예: 95% 신뢰수준)
confidence_level = 0.95
alpha = 1 - confidence_level

# t-분포의 임계값 계산
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# 신뢰구간 계산
margin_of_error = t_critical * (sample_std / np.sqrt(n))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"신뢰구간: {confidence_interval}")
```

```python
import matplotlib.pyplot as plt
from scipy.stats import binom
import koreanize_matplotlib
```

```python
# 이항분포
# 고정된 횟수의 독립적인 시행에서 두 가지 가능한 결과(성공 또는 실패)를 가지는 확률 분포입니다.
# 각 시행은 독립적, 각 시행은 같은 확률
# 성공 / 실패 , 이진 결과를 가진 실험에서 사용
```

```python
# 이항분포 매개변수
n = 10  # 시행 횟수
p = 0.5  # 성공 확률

# x 값 (성공 횟수)
x = np.arange(0, n + 1)

# 확률 질량 함수 (PMF)
pmf_binom = binom.pmf(x, n, p)

# 시각화
plt.figure(figsize=(10, 5))
plt.bar(x, pmf_binom, color='blue', alpha=0.7, label='PMF')
plt.xlabel('성공 횟수 (k)')
plt.ylabel('확률 P(X=k)')
plt.title(f'이항분포 (n={n}, p={p})')
plt.legend()
plt.show()

# 이항분포에 따른 랜덤 샘플 생성
sample_size = 1000
samples_binom = np.random.binomial(n, p, sample_size)

# 랜덤 샘플의 히스토그램 시각화
plt.figure(figsize=(10, 5))
plt.hist(samples_binom, bins=np.arange(0, n + 2) - 0.5, edgecolor='black', alpha=0.7, color='blue')
plt.xlabel('성공 횟수 (k)')
plt.ylabel('빈도수')
plt.title(f'이항분포에서 추출한 랜덤 샘플 (n={n}, p={p}, 샘플 크기={sample_size})')
plt.show()
```

```python
# 포아송
# 일정한 시간 또는 공간 내에서 사건이 발생하는 횟수를 모델링
# 사건은 독립적, 평균 발생률은 같다.
# 특정시간 동안 발생하는 사건의 수.
```

```python
from scipy.stats import poisson

# 포아송분포 매개변수
lambda_poisson = 5  # 평균 사건 발생 횟수

# x 값 (사건 횟수)
x_poisson = np.arange(0, 20)

# 확률 질량 함수 (PMF)
pmf_poisson = poisson.pmf(x_poisson, lambda_poisson)

# 시각화
plt.figure(figsize=(10, 5))
plt.bar(x_poisson, pmf_poisson, color='green', alpha=0.7, label='PMF')
plt.xlabel('사건 횟수 (k)')
plt.ylabel('확률 P(X=k)')
plt.title(f'포아송분포 (λ={lambda_poisson})')
plt.legend()
plt.show()

# 포아송분포에 따른 랜덤 샘플 생성
sample_size_poisson = 1000
samples_poisson = np.random.poisson(lambda_poisson, sample_size_poisson)

# 랜덤 샘플의 히스토그램 시각화
plt.figure(figsize=(10, 5))
plt.hist(samples_poisson, bins=np.arange(0, max(samples_poisson) + 2) - 0.5, edgecolor='black', alpha=0.7, color='green')
plt.xlabel('사건 횟수 (k)')
plt.ylabel('빈도수')
plt.title(f'포아송분포에서 추출한 랜덤 샘플 (λ={lambda_poisson}, 샘플 크기={sample_size_poisson})')
plt.show()
```
