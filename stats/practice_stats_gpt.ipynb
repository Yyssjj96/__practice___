- 귀무가설과 대립가설: 가설 검정은 귀무가설(Null Hypothesis)과 대립가설(Alternative Hypothesis)을 설정하여 특정 통계적 검정을 수행하는 것입니다. 귀무가설은 일반적으로 "효과가 없다" 또는 "차이가 없다"라는 가정을, 대립가설은 반대의 가정을 설정합니다.
유의수준과 p-value:

- 유의수준: 가설 검정에서 오류를 감수할 확률을 나타내는 값입니다. 보통 0.05 또는 0.01이 사용됩니다.
- p-value: 귀무가설이 참이라는 가정 하에 얻은 검정 통계량보다 극단적인 결과를 얻을 확률입니다. 작을수록 귀무가설을 기각할 근거가 더 강력합니다.

- 통계적 검정 방법:
t-test: 평균 차이를 비교할 때 사용됩니다. 독립 표본 t-test와 대응 표본 t-test로 나뉩니다.
ANOVA(Analysis of Variance): 세 개 이상의 그룹 간의 평균 차이를 비교할 때 사용됩니다. 일원배치 ANOVA와 이원배치 ANOVA 등이 있습니다.
카이제곱 검정(Chi-square test): 범주형 자료의 독립성을 검정할 때 사용됩니다.
회귀 분석: 변수들 간의 관계를 분석할 때 사용됩니다.
오차의 종류:

- 제1종 오류(Type I Error): 귀무가설이 참일 때 이를 잘못 기각하는 오류입니다. 유의수준과 관련이 있습니다.
- 제2종 오류(Type II Error): 귀무가설이 거짓일 때 이를 잘못 채택하는 오류입니다. 표본 크기와 관련이 있습니다.
표본 크기와 검정력:

- 표본 크기가 클수록 통계적 검정의 검정력이 증가하며, 작은 차이도 통계적으로 유의미하게 될 가능성이 높아집니다.
- 비모수적 방법:
데이터가 정규분포를 따르지 않을 때 사용할 수 있는 검정 방법입니다. 예를 들어, 윌콕슨 순위합 검정(Wilcoxon Rank-Sum Test)이나 크루스칼-왈리스 검정(Kruskal-Wallis Test) 등이 있습니다.
다중 검정 보정:

- 정규성 검정(Shapiro-Wilk Test 등)과 등분산성 검정(Bartlett Test 등)은 가설 검정 전에 데이터의 전제 조건을 확인하는 데 중요합니다.

따라서 비모수적 방법은 데이터 분포의 특성을 덜 고려하고, 보다 일반적인 상황에서 통계적 검정을 수행할 수 있는 장점이 있습니다.

Shapiro-Wilk 정규성 검정은 주로 다음과 같은 경우에 사용됩니다:

- 통계적 가정 확인: 다른 통계적 분석을 수행하기 전에 데이터가 정규분포를 따르는지 확인할 때 사용됩니다. 대표적으로 t-test, 분산분석(ANOVA) 등의 분석을 수행하기 전에 정규성 검정을 진행합니다.

- 데이터의 특성 파악: 데이터의 분포 특성을 파악하고자 할 때 사용됩니다. 데이터의 분포가 정규분포에 가까운지, 혹은 비정규적인 특성을 보이는지 판단하는 용도로 활용됩니다.

- 표본의 크기 결정: 표본의 크기를 결정할 때, 정규성 검정을 통해 표본이 충분히 정규분포를 따르는지를 확인할 수 있습니다. 특히 통계적 추정이나 신뢰구간 계산 시 정규성 가정이 필요한 경우가 많습니다.

- 이상치 탐지: 데이터에서 이상치가 있는지를 확인할 때, 정규성 검정을 통해 이상치가 데이터의 정규성을 크게 벗어나는지를 판단할 수 있습니다.

따라서 Shapiro-Wilk 정규성 검정은 데이터 분석의 첫 단계에서 데이터의 분포 특성을 확인하고, 이후 분석 방법을 선택하거나 데이터의 해석에 필요한 사전 조건으로 사용됩니다.

```python
# Example of the Shapiro-Wilk Normality Test
from scipy.stats import shapiro
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably Gaussian')
else:
 print('Probably not Gaussian')
 # 귀무 : 정규분포를 따른다
 # 대립 : 정규분포를 따르지 않는다.

 # p > 0.05 = 정규분포를 따른다.
```

```python
# Example of the Pearson's Correlation test
from scipy.stats import pearsonr
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

```python
# Example of the Pearson's Correlation test
from scipy.stats import pearsonr
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
 print('Probably independent')
else:
 print('Probably dependent')
```

- 등분산성(等變數性, homogeneity of variances)은 통계학에서 중요한 개념으로, 두 개 이상의 모집단이나 집단에서 변량(변수)의 분산이 서로 동일한지를 의미합니다. 다시 말해, 서로 다른 집단들에서 측정한 변수의 분산이 일정하다는 것을 말합니다.

등분산성의 중요성
등분산성은 주로 다음과 같은 상황에서 중요성을 가집니다:

- 독립표본 t-test: 두 개의 독립적인 집단에서 평균 차이를 비교할 때, 두 집단의 분산이 같다는 가정이 필요합니다. 만약 등분산성이 만족되지 않으면, t-test의 신뢰도가 떨어질 수 있습니다.

- 분산 분석(ANOVA): 세 개 이상의 그룹 간 평균 차이를 비교할 때, 각 그룹의 분산이 서로 같은지 확인하는 것이 중요합니다. 등분산성이 없으면 ANOVA 결과의 정확성이 떨어질 수 있습니다.

- 회귀 분석: 다중 선형 회귀 모델에서 독립 변수들 간에 분산이 일정하다는 가정이 필요합니다. 이를 위해 등분산성을 확인하는 것이 중요합니다.

등분산성 검정 방법
등분산성을 확인하기 위한 일반적인 검정 방법으로는 다음과 같은 것들이 있습니다:

- Levene의 검정(Levene's test): 각 집단에서 관측값들의 분산이 동일한지를 검정하는 방법입니다. 이 방법은 비모수적인 방법으로서, 데이터가 정규분포를 따르지 않아도 상대적으로 강건한 특성을 가집니다.

- Bartlett의 검정(Bartlett's test): 각 집단에서 관측값들이 정규분포를 따르고 있을 때 사용하는 등분산 검정 방법입니다. 정규성을 가정하고 있기 때문에, 데이터가 정규분포를 따를 때 유용합니다.

- Brown-Forsythe 검정: Levene의 검정과 비슷하지만, 중앙값(median)을 기준으로 분산의 차이를 검정합니다. 이 방법도 비모수적이며, 정규성 가정이 필요하지 않습니다.

결론
등분산성은 다양한 통계적 분석에서 중요한 전제 조건 중 하나입니다. 데이터 분석에서 등분산성을 확인하는 것은 결과의 신뢰성을 높이는 데 큰 도움이 됩니다. 따라서, 분석을 시작하기 전에 각 집단의 분산이 동일한지를 확인하는 것이 좋습니다.
