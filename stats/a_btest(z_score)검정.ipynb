```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ab_data.csv')
```

```python
df.head(15)
```

```python
# 가설
# 귀무 : 두 그룹의 전환율은 차이가없다.
# 대립 : 두 그룹의 전환율은 차이가있다.
```

```python
control_conversion_rate = df[df['group'] == 'control']['converted'].mean()
treatment_conversion_rate = df[df['group'] == 'treatment']['converted'].mean()
```

```python
# 각 그룹의 전환된 사용자 수와 전환되지 않은 사용자 수 계산
control_converted = df[df['group'] == 'control']['converted'].sum()
control_total = df[df['group'] == 'control']['converted'].count()
treatment_converted = df[df['group'] == 'treatment']['converted'].sum()
treatment_total = df[df['group'] == 'treatment']['converted'].count()

# 전환율 차이의 표준 오차 계산
pooled_conversion_rate = (control_converted + treatment_converted) / (control_total + treatment_total)
standard_error = np.sqrt(pooled_conversion_rate * (1 - pooled_conversion_rate) * (1/control_total + 1/treatment_total))
```

```python
z_score = (treatment_conversion_rate - control_conversion_rate) / standard_error
```

```python
from scipy import stats
p_value = stats.norm.sf(abs(z_score)) * 2  # 양측 검정
```

```python
# 결과 출력
print("Control Conversion Rate:", control_conversion_rate)
print("Treatment Conversion Rate:", treatment_conversion_rate)
print("Z-Score:", z_score)
print("P-Value:", p_value)
```
