```python
from konlpy.tag import Okt

okt = Okt()

# 'content' 열에서 형태소 분석 후 명사, 동사, 형용사만 추출하는 함수 정의
def extract_noun_verb_adj(text):
    pos_list = okt.pos(text)
    result = [word for word, pos in pos_list if pos in ['Noun']]
    return result

# 'content' 열에 대해 형태소 분석 및 명사, 동사, 형용사 결과 저장
df_negative['na_words'] = df_negative['content'].apply(extract_noun_verb_adj)

# 결과 확인
print(df_negative['na_words'])
```

```python
from collections import Counter

exclude_words = ['치', '좀', '직', '앱','때','수','것','트위치','기능','왜','창','방송']

# 'na_words' 열에서 모든 명사들을 하나의 리스트로 모음 (제외 단어 필터링)
all_nouns = []
for words_list in df_negative['na_words']:
    filtered_words = [word for word in words_list if word not in exclude_words]
    all_nouns.extend(filtered_words)

# 명사들의 등장 횟수를 카운트
word_counts = Counter(all_nouns)

# 가장 많이 나온 단어 10개 출력
top_words = word_counts.most_common(10)
print("가장 많이 나온 단어들 (제외 단어 포함):")
for word, count in top_words:
    print(f"{word}: {count}번")
```
